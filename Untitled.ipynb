```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option("display.max_columns",None)
```

```python
data=pd.read_csv("C:/users/daniel/documents/edureka/house prices/train.csv")
```

```python
data.head()
```

```python
data=data.drop(columns="Id")
```

#### Features that contain Null values

```python
null_feat=[i for i in data.columns if data[i].isnull().sum()>0]
null_feat
```

```python
print(data[null_feat].shape)
data[null_feat].head()
```

```python
#### The number of null values in each feature

data[null_feat].isnull().sum()
```

#### The number of null_features that are Numerical

```python
num_null_feat=[i for i in null_feat if data[i].dtypes !="O"]

print(data[num_null_feat].shape)
data[num_null_feat].head()
```

```python
data[num_null_feat].isnull().sum()
```

#### The number of null_features that are Categorical

```python
cat_null_feat=[i for i in null_feat if i not in num_null_feat]

print(data[cat_null_feat].shape)
data[cat_null_feat].head()
```

```python
data[cat_null_feat].isnull().sum()
```

```python
#### Let us get the number of unique values in the cat null features

no=[(i,len(data[cat_null_feat][i].unique())) for i in data[cat_null_feat]]
no
```

```python
#### Let us get the sum of the unqiue values in each cat_null_feat
```

```python
countval=[(i,data[i].value_counts(dropna=False)) for i in data[cat_null_feat]]

countval
```

### Non-null features

```python
non_nullfeat=[i for i in data.columns if i not in null_feat]

print(data[non_nullfeat].shape)
data[non_nullfeat].head()
```

```python
#### numerial non_null

num_nonnull=[i for i in non_nullfeat if data[i].dtypes != "O"]

print(data[num_nonnull].shape)
data[num_nonnull].head()
```

```python
#### cat non null

cat_nonnull=[i for i in non_nullfeat if i not in num_nonnull]

print(data[cat_nonnull].shape)
data[cat_nonnull].head()
```

```python
#### Let us count the number of unique values in each categorical non null feature

num_var=[(i, data[i].value_counts()) for i in data[cat_nonnull]]
num_var
```
